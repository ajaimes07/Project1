```python
import pandas as pd
import numpy as np
import matplotlib as plt

csv_path = "Housepy_raw.csv"
csv_path2="HouseHold income 2013.2017.csv"
housepy_df = pd.read_csv(csv_path)
housepy_df.head()
householdincome = pd.read_csv(csv_path2)
householdincome.head()
housepy_df.head()
```

```python
#dropping columns that wont be used to tell us anything

housepy_clean = housepy_df[["MLS Number", "Street Number", "Street Name", "City/Location", "Zip Code", "List Price",
                           "Close Price", "Close Date", "Area", "Subdivision", "Market Area", "School District",
                           "School Elementary", "School Middle", "School High", "Sq Ft Total", "Price Sq Ft Sold", "Lot Size",
                          "Year Built", "Bedrooms", "Baths Total", "New Construction", "Pool Private", "No Of Garage Cap",
                          "Style", "CDOM"]]
householdincome_clean=householdincome[["Year","Household Income by Race","Geography"]]
housepy_clean.head(5)
housepy_clean['Close Price'] = housepy_clean['Close Price'].apply(lambda x: x.replace("$", ""))
housepy_clean['Close Price'] = housepy_clean['Close Price'].apply(lambda x: x.replace(",", ""))
housepy_clean['Close Price'] = housepy_clean['Close Price'].astype(int)
housepy_clean.head()
len(housepy_clean)
```
